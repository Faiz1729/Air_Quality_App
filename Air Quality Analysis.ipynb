### Data Preprocessing and Cleaning

```python
new_df.dtypes
```

```python
def rep(par):
    m = round(par.mean(),2)
    
    l = []
    for i in par:
        if i == 0.00:
            l.append(m)
        else:
            l.append(i)
    
    return l        
```

```python
new_df["PM2.5"] = rep(new_df["PM2.5"])
new_df
```

```python
new_df["PM2.5"] = rep(new_df["PM2.5"])
new_df["SO2"] = rep(new_df["SO2"])
new_df["NOx"] = rep(new_df["NOx"])
new_df["PM10"] = rep(new_df["PM10"])

new_df["NH3"] = rep(new_df["NH3"])
```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'])
```

```python
per = new_df['Date'].dt.to_period("M")
g = new_df.groupby(per)
monthlydf = round(g.mean(),2)
monthlydf
```
